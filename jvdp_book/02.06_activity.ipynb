```python
%matplotlib inline

import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt

file = fits.open('Data/UDF_00347.2D.fits')
plt.imshow(file[1].data, cmap='viridis',vmin=0, vmax=0.25)

mask = file[3].data == 347

total_es = np.sum(file[1].data[mask])

f_lambda =  total_es*1.4737e-20          
print('Flux: ', f_lambda)
                               
#f_nu = 3.34e4*(lambda**2)*f_lambda
f_nu = 3.34e4*(13923.**2)*f_lambda*1e-23
AB_mag = -2.5*np.log10(f_nu) - 48.6
                               
print('AB mag: ', AB_mag)

print(file[0].header['MAG'], file[0].header['FTOTAL'])
```
