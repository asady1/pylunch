`astropy.table` automatically knows what to do with astronomy formats. E.g. ApJ tables:

```python
sntab = Table.read("ftp://cdsarc.u-strasbg.fr/pub/cats/VII/253/snrs.dat",
                   readme="ftp://cdsarc.u-strasbg.fr/pub/cats/VII/253/ReadMe",
                   format='ascii.cds')
```

[DECaLS](http://legacysurvey.org/) data for [the field around the Leo I dwarf](http://legacysurvey.org/viewer?ra=152.1170833&dec=12.3063889&layer=decals-dr3) as a *fits* table:

```python
leoidecals = Table.read('http://portal.nersc.gov/project/cosmo/data/legacysurvey/dr3/tractor/152/tractor-1520p122.fits')
```

Of course it can turn them into pandas tables:

```python
sntab_df = sntab.to_pandas()
type(sntab_df)
```

But where's the fun in that?  What can it do differently?

```python
sntab_df
```

```python
sntab
```

Some immediate apparent differences:

* Astropy does not have an (expplicit) "index" column
* Astropy handles "missing" data differently.
* Astropy columns have units

### Indexing 

This underlies an important conceptual distinction: an astropy table is what it says: a *table*, not a "data frame".  So in general it behaves like a table, with columns indexed on names and rows indexed by row number:

```python
sntab[0]
```

```python
sntab['S(1GHz)']
```

The row access can also slice up tables trivially with numpy-style indexing:

```python
sntab[sntab['MajDiam']<5]
```

In contrast to pandas, which supports column access but requires more indirection to access rows:

```python
sntab_df['MajDiam']
```

```python
sntab_df[0]
```

```python
sntab_df.loc[0]
```

###  Missing data

Astropy tables uses masked columns for missing data rather than using NaN's.  This provides a more consistent experience for non-float dtypes (pandas has to cast them to float to support NaN's).

```python
sntab['MinDiam']
```

```python
sntab_df['MinDiam']
```

You can interact with the astropy version just like you can with numpy masked arrays:

```python
sntab['MinDiam'].filled(-1)
```

```python
sntab[~sntab['MinDiam'].mask]
```

### Units and "mixin" columns

One of the biggest differences in astropy.table is that the "columns" are more flexible in their behavior.  For example, they can behave as unitful Quantities if compared with other Quantities

```python
from astropy import units as u
```

```python
np.sum(u.Quantity(leoidecals['shapeExp_r']) > 0.5*u.arcmin) / len(leoidecals)
```

Moreover, just like with numpy structured arrays, astropy table elements can be more complex things.  For example, they could be multi-dimensional arrays:

```python
leoidecals['decam_flux'].shape
```

```python
mags = 22.5-2.5*np.log10(leoidecals['decam_flux'])
g = mags[:, 1]
r = mags[:, 2]

plt.scatter(g-r, r, alpha=.2, s=2)
plt.xlim(-0.5, 2)
plt.ylim(24, 15)
```

They can even be astropy objects like coordinates:

```python
from astropy.coordinates import SkyCoord
leoidecals['coords'] = SkyCoord(leoidecals['ra'], leoidecals['dec'])
```

```python
sep = leoidecals['coords'].separation(SkyCoord.from_name('Leo I '))

msk = sep.arcmin < 5
plt.scatter((g-r)[msk], r[msk], alpha=.2, s=2)
plt.xlim(-0.5, 2)
plt.ylim(24, 15)
```
